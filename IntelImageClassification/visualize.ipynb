```python
import matplotlib.pyplot as plt
import scienceplots

plt.style.use('science')
plt.style.use('ggplot')
#plt.rcParams["font.family"] = "Arial"
#plt.rcParams["font.weight"] = "bold"
#plt.rcParams["axes.labelweight"] = "bold"

def plotLoss(training_loss, val_loss):
    cm = 1/2.54
    fig1 = plt.figure(figsize=(15*cm,10*cm), dpi=300)
    ax = fig1.subplots(1)
    ax.plot(range(1,21), training_loss, label='Training loss', marker='^', c='#D35400', markersize=8, linewidth = 0.8)
    ax.plot(range(1,21), val_loss, label='Validation loss', marker='D', c='#7D3C98', markersize=8, linewidth = 0.8)
    ax.set_xticks(range(5,21,5))
    plt.xlabel('Epoch', fontsize=15)
    plt.ylabel('Loss', fontsize=15)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.legend(prop = {'size' : 15})
    plt.show()

def plotGE(GE1, GE2):
    cm = 1/2.54
    fig1 = plt.figure(figsize=(15*cm,10*cm), dpi=300)
    ax = fig1.subplots(1)
    ax.plot(range(1,21), GE1, label='CE', c='b', marker='o', markersize=7, linewidth = 0.8)
    ax.plot(range(1,21), GE2, label='RJM', c='r', marker='s', markersize=7, linewidth = 0.8)
    ax.set_xticks(range(5,21,5))
    plt.xlabel('Epoch', fontsize=15)
    plt.ylabel('$\hat{E}(f_{B_S,R})$', fontsize=15)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.legend(prop = {'size' : 15})
    plt.show()
    
```
